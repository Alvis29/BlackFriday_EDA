```python
# group up the data by user_id
grouped_data = df[['User_ID','Gender', 'Age', 'Occupation',
       'City_Category', 'Stay_In_Current_City_Years', 'Marital_Status']].groupby(by='User_ID').first()
```

### univariate analysis

```python
ax = sns.countplot(data=grouped_data , x='Gender')
for p in ax.patches:
    ax.annotate(f'\n{p.get_height()}', (p.get_x()+0.4, p.get_height()), ha='center', va='top', color='black')

plt.title('count of number female and male shop during balck friday')
plt.show()
```

count of males who did shopping are more than 2.5 times count of female



```python
sns.countplot(data=grouped_data , x='Age', order = [x for x in grouped_data.Age.value_counts().sort_index().index])

plt.title('Age group classification of shopping')
plt.show()
```

```python
# Data
labels = [x for x in grouped_data.Age.value_counts().sort_index().index]
sizes = [x for x in grouped_data.Age.value_counts().sort_index().values]

# Create a circle for the center of the chart
circle = plt.Circle((0, 0), 0.7, color='white')

# Create a pie chart with a hole in the center
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, labeldistance=1.05)
ax.add_artist(circle)

# Add a title
ax.set_title('Age distribution')

# Display the chart
plt.show()
```

age group 18 - 45 consist most of the customers,to be precise 72.7% of total customers lie in the age group.

and perticularly age group 26 -35 are the one consists most number of customers

```python
sns.countplot(grouped_data.Occupation)

plt.title('density of people shopping with respect tp working years')
plt.show()
```

```python
temp = grouped_data.City_Category.value_counts().sort_index()

categories = [x for x in temp.index]
counts = [x for x in temp.values]

sns.set_style("whitegrid")


max_index = counts.index(max(counts))

explosion = [0, 0, 0]
explosion[max_index] = 0.1

plt.figure(figsize=(6, 6))
wedges, labels, _ = plt.pie(counts, labels=categories, explode=explosion, autopct='%1.1f%%', startangle=90, shadow=True)


wedges[max_index].set_edgecolor('black')
wedges[max_index].set_alpha(0.8)

plt.title("percentage of people who did shopping wrt given city")
plt.legend()


plt.show()
```

```python
print('total money spent by female customers are Rs.',spending_by_category[0])
print('total money spent by male customers are Rs.',spending_by_category[1])
print('this is obvivious as number of male customers are more than 2.5 times than female customers')

mean_spend_female = round(spending_by_category[0]/grouped_data['Gender'].value_counts()[1],2)
mean_spend_male = round(spending_by_category[1]/grouped_data['Gender'].value_counts()[0],2)

print()
print('average money spent by a female customer are Rs.',mean_spend_female)
print('average money spent by a male  customer are Rs.',mean_spend_male)
```
